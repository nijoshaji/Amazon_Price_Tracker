```python
## create a time stamp for the extraction. YYYY-MM-DD format

date_stamp = datetime.date.today()
#print(date_stamp)
```

```python
## create a csv and write the info into it. Using csv library

import csv

# prepare the data
header = ['Title','Price','Rating','Date']
row_data = [title, price, rating, date_stamp]


with open ("price.csv",'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(row_data)
```
